```python
# Author:       Rick Candell (rick.candell@nist.gov)
#               Tim Zimmerman (timothy.zimmerman@nist.gov)
# Organization: National Institute of Standards and Technology
#               U.S. Department of Commerce
# License:      Public Domain

import matplotlib.pyplot as plt
import numpy as np
import csv

fname = r'C:/Users/taz2/Projects/tesim/c/teplant.dat'
p = np.loadtxt(fname, delimiter="\t")
fname = r'C:/Users/taz2/Projects/tesim/c/tectlr.dat'
c = np.loadtxt(fname, delimiter="\t")
fname = r'C:/Users/taz2/Projects/tesim/c/ricker_baseline.dat'
b = np.loadtxt(fname, delimiter="\t")

# Defines the start time of the plots
tstart = 0.0 # hours
# Defines the span of time (in hours) of the plots
tspan = 4.0 # hours
# Defines the delta t of data acquisition
delta_t = 0.01 # hours

tstart *= 1. / delta_t
tspan *= 1. / delta_t
```

```python
xmv_names = [
'D Feed',
'E Feed',
'A Feed',
'A+C Feed',
'Recycle Valve',
'Purge Valve',
'Separator Valve',
'Stripper Valve',
'Steam valve',
'Reactor Coolant',
'Condenser Coolant',
'Agitator Speed' ]

xmeas_names = [
'A Feed (kscmh)',
'D Feed (kg/hr)',
'E Feed (kg/hr)',
'A+C Feed (kscmh)',
'Recycle Flow (kscmh)',
'Reactor Feed (kscmh)',
'Reactor Pressure (kPa)',
'Reactor Level (%)',
'Reactor Temp. (°C)',
'Purge Rate (kscmh)',
'Sep Temp. (°C)',
'Sep Level (%)',
'Sep Pressure (kPa)',
'Sep Underflow (m³/hr)',
'Stripper Level (%)',
'Stripper Pressure (kPa)',
'Stripper Underflow (m³/hr)',
'Stripper Temp. (°C)',
'Steam Flow (kg/hr)',
'Compressor Work (kW)',
'React. Cool. Temp. (°C)',
'Cond. Cool. Temp. (°C)',
'Feed A (mole %)',
'Feed B (mole %)',
'Feed C (mole %)',
'Feed D (mole %)',
'Feed E (mole %)',
'Feed F (mole %)',
'Purge     % A (mole %)',
'Purge     % B (mole %)',
'Purge     % C (mole %)',
'Purge     % D (mole %)',
'Purge     % E (mole %)',
'Purge     % F (mole %)',
'Purge     % G (mole %)',
'Purge     % H (mole %)',
'Product D (mole %)',
'Product E (mole %)',
'Product F (mole %)',
'Product G (mole %)',
'Product H (mole %)' ]
```

```python
# Plot all measured variables as seen by the plant and compare
# to the baseline for each variable.

plt.figure(figsize=(16,50))
tp=np.array(p[tstart:(tstart+tspan),0])
tc=np.array(c[tstart:(tstart+tspan),0])
tb=np.array(b[tstart:(tstart+tspan),0])
for ii in range(0,41):
    ind = ii+13
    plt.subplot(14,3,ii+1)
    
     # Plant Simulation Plot (BLUE)
    plt.plot(tp,p[tstart:(tstart+tspan),ind],linewidth=0.5,marker="o",color="b")
    
    # Controller Simulation Plot (GREEN)
    plt.plot(tc,c[tstart:(tstart+tspan),ind],linewidth=0.5,marker="o",color="g")
    
    # Baseline Plot (RED)
    plt.plot(tb,b[tstart:(tstart+tspan),ind],linewidth=0.5,marker=".",color="r")
    
    plt.title(xmeas_names[ii])
```

```python
# Plot measured variables as seen by the controller
# NOTE: This plot does NOT compare with the plant. It compares
#       controller xmv values vs. expected xmv values (baseline)

plt.figure(figsize=(20,16))
tp=np.array(p[tstart:(tstart+tspan),0])
tc=np.array(c[tstart:(tstart+tspan),0])
tb=np.array(b[tstart:(tstart+tspan),0])
for ii in range(0,12):
    ind = ii+1
    plt.subplot(4,3,ii+1)
    
    # Controller Simulation Plot (GREEN)
    plt.plot(tc,c[tstart:(tstart+tspan),ind],linewidth=0.5,marker="o",color="g")
    
    # Baseline Plot (RED)
    plt.plot(tb,b[tstart:(tstart+tspan),ind],linewidth=0.5,marker=".",color="r")
    
    plt.title(xmv_names[ii])
```

```python
# Plots the production rate and mole % of production

# This plot has its own prod_tstart and prod_tspan because of
# the limited rate of capture vs. other variables

# Defines the start time (in hours) of the plots
prod_tstart = 0.0
# Defines the span of time (in hours) of the plots
prod_tspan = 72.0

prod_tstart *= 100.
prod_tspan *= 100.

plt.figure(figsize=(20,5))
tc=np.array(c[prod_tstart:(prod_tstart+prod_tspan),0])
plt.plot(tc,c[prod_tstart:(prod_tstart+prod_tspan),29],linewidth=0.5)
plt.title('Production Rate')
plt.xlim(prod_tstart,prod_tspan/100)
plt.grid(b=True, which='major')

plt.figure(figsize=(20,10))
plt.plot(tc,c[prod_tstart:(prod_tstart+prod_tspan),49],linewidth=1.0,label='mole % of D in Product')
plt.plot(tc,c[prod_tstart:(prod_tstart+prod_tspan),50],linewidth=1.0,label='mole % of E in Product')
plt.plot(tc,c[prod_tstart:(prod_tstart+prod_tspan),51],linewidth=1.0,label='mole % of F in Product')
plt.plot(tc,c[prod_tstart:(prod_tstart+prod_tspan),52],linewidth=1.0,label='mole % of G in Product')
plt.plot(tc,c[prod_tstart:(prod_tstart+prod_tspan),53],linewidth=1.0,label='mole % of H in Product')
plt.grid(b=True, which='major')
plt.legend(loc='lower right')
plt.xlim(prod_tstart,prod_tspan/100)
plt.title('mole % of Production')
```
